## **Tokenization**

```python

from nltk.tokenize import word_tokenize, TreebankWordTokenizer, RegexpTokenizer, WhitespaceTokenizer, TweetTokenizer, PunktSentenceTokenizer, sent_tokenize
from nltk.corpus import gutenberg



class NLPTokenizer:
    def __init__(self, text, tokenizer_type,isLower=True):
        self.text = text
        self.tokenizer_type = tokenizer_type
        self.isLower = isLower
    def process(self):
      if self.isLower:
        self.text=self.text.lower()
    def tokenizer(self):
        self.process()
        if self.tokenizer_type == "word_tokenize":
            return word_tokenize(self.text)

        elif self.tokenizer_type == "treebank":
            tokenizer = TreebankWordTokenizer()
            return tokenizer.tokenize(self.text)

        elif self.tokenizer_type == "regexp":
            tokenizer = RegexpTokenizer(r'\w+')
            return tokenizer.tokenize(self.text)

        elif self.tokenizer_type == "whitespace":
            tokenizer = WhitespaceTokenizer()
            return tokenizer.tokenize(self.text)

        elif self.tokenizer_type == "tweet":
            tokenizer = TweetTokenizer()
            return tokenizer.tokenize(self.text)

        elif self.tokenizer_type == "sent_tokenize":
            return sent_tokenize(self.text)

        elif self.tokenizer_type == "punkt":
            tokenizer = PunktSentenceTokenizer()
            return tokenizer.tokenize(self.text)

        else:
            raise ValueError(f"Unknown tokenizer type: {self.tokenizer_type}")

    def print_tokens(self):
        tokens = self.tokenizer()
        print(f"\n--- Tokens using {self.tokenizer_type} ---")
        print(tokens)
        print("-"*200)


sample_text = gutenberg.raw('austen-emma.txt')[:300]


tokenizer_types = ["word_tokenize", "treebank", "regexp", "whitespace", "tweet", "sent_tokenize", "punkt"]
print("-"*200)
print("With lower casing ")
print("-"*200)
for t_type in tokenizer_types:
    tokenizer = NLPTokenizer(sample_text, t_type)
    tokenizer.print_tokens()
print("-"*200)
print("Without lower casing")
print("-"*200)
for t_type in tokenizer_types:
    tokenizer = NLPTokenizer(sample_text, t_type,False)
    tokenizer.print_tokens()
```

# **Lemmatization and stemming**

```python
import nltk
import string
from nltk.corpus import stopwords,wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer

class StopwordRemover:
  def __init__(self,text):
    self.text=text

  def remove_stopwords(self):
    tokenizer=NLPTokenizer(self.text,'word_tokenize',False)
    tokens=tokenizer.tokenizer()
    stop_words=set(stopwords.words('english'))
    filtered_tokens=[word for word in tokens if word.lower() not in stop_words]
    return filtered_tokens
  def print_output(self):
    output=self.remove_stopwords()
    print(f"\n--- Tokens after removing stopwords ---")
    print(output)
    print("-"*200)

sample_text = "Emma, by Jane Austen (1816): Emma Woodhouse, handsome, clever, and rich, had a happy disposition."


stopword_remover = StopwordRemover(sample_text)
stopword_remover.print_output()
```

```python
class Stemmer:
  def __init__(self,text):
    self.text=text

  def stem_words(self):
    tokenizer=NLPTokenizer(self.text,'word_tokenize',False)
    tokens=tokenizer.tokenizer()
    stemmer=PorterStemmer()
    stemmed_tokens=[stemmer.stem(word) for word in tokens]
    return stemmed_tokens
  def print_output(self):
    output=self.stem_words()
    print(f"\n--- Tokens after stemming ---")
    print(output)
    print("-"*200)

sample_text = "Emma, by Jane Austen (1816): Emma Woodhouse, handsome, clever, and rich, had a happy disposition."


stemmer = Stemmer(sample_text)
stemmer.print_output()
```

```python
class Lemmatizer:
  def __init__(self,text):
    self.text=text
  def lemmatize_words(self):
    tokenizer=NLPTokenizer(self.text,'word_tokenize',False)
    tokens=tokenizer.tokenizer()
    lemmatizer=WordNetLemmatizer()
    lemmatized_tokens=[lemmatizer.lemmatize(word) for word in tokens]
    return lemmatized_tokens
  def print_output(self):
    output=self.lemmatize_words()
    print(f"\n--- Tokens after lemmatization ---")
    print(output)
    print("-"*200)

sample_text = "Emma, by Jane Austen (1816): Emma Woodhouse, handsome, clever, and rich, had a happy disposition."
lemmatizer = Lemmatizer(sample_text)
lemmatizer.print_output()
```

```python
class PunctuationRemover:
  def __init__(self,text):
    self.text=text

  def remove_punctation(self):
    translator=str.maketrans('','',string.punctuation)
    no_punct_text=self.text.translate(translator)
    return no_punct_text
  def print_output(self):
    output=self.remove_punctation()
    print(f"\n--- Text after removing punctuation ---")
    print(output)
    print("-"*200)
sample_text = "Emma, by Jane Austen (1816): Emma Woodhouse, handsome, clever, and rich, had a happy disposition."
punctuation_remover = PunctuationRemover(sample_text)
punctuation_remover.print_output()
```

# **Feature Engineering- Bag of words, TF-IDF, Word Embeddings and Sentence Embeddings**

```python
import nltk
from sklearn.feature_extraction.text import TfidfTransformer,CountVectorizer
from gensim.models import Word2Vec
from sentence_transformers import SentenceTransformer

class BagOfWordsVectorizer:
  def __init__(self,text):
    self.text=text
    self.vectorizer=CountVectorizer()

  def transform(self):
    bow_matrix=self.vectorizer.fit_transform(self.text)
    return bow_matrix.toarray(),self.vectorizer.get_feature_names_out()

  def print_output(self):
    bow_matrix,feature_names=self.transform()
    print(f"\n--- Bag of Words Matrix ---")
    print(bow_matrix)
    print("\n--- Feature Names ---")
    print(feature_names)
    print("-"*200)
sample_corpus = [
    "Emma Woodhouse is handsome, clever, and rich.",
    "She had lived nearly twenty-one years in the world with very little to distress or vex her."
]
bow_vectorizer=BagOfWordsVectorizer(sample_corpus)
bow_vectorizer.print_output()
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

class TFIDFVectorizer:
    def __init__(self, corpus):
        self.corpus = corpus
        self.vectorizer = TfidfVectorizer()

    def transform(self):

        corpus = np.array(self.corpus).reshape(-1, 1) if len(self.corpus) == 1 else self.corpus
        tfidf_matrix = self.vectorizer.fit_transform(corpus)
        return tfidf_matrix.toarray(), self.vectorizer.get_feature_names_out()

    def print_output(self):
        matrix, feature_names = self.transform()
        print("\n--- TF-IDF ---")
        print("Feature Names:", feature_names)
        print("Matrix:\n", matrix)


sample_corpus = [
    "Emma Woodhouse is handsome, clever, and rich.",
    "She had lived nearly twenty-one years in the world with very little to distress or vex her."
]

tfidf_vectorizer = TFIDFVectorizer(sample_corpus)
tfidf_vectorizer.print_output()
```

```python
class WordEmbeddingsVectorizer:
  def __init__(self,corpus):
    self.corpus=[nltk.word_tokenize(doc.lower()) for doc in corpus]
  def train(self):
      model = Word2Vec(sentences=self.corpus, vector_size=100, window=5, min_count=1, workers=2, sg=0)
      return model
  def print_output(self):
      model = self.train()
      print("\n--- Word Embeddings (Word2Vec) ---")
      print("Vocabulary:", list(model.wv.index_to_key))
      print("Embedding for 'emma' (if available):", model.wv['emma'] if 'emma' in model.wv else "Not Found")

word_embeddings_vectorizer = WordEmbeddingsVectorizer(sample_corpus)
word_embeddings_vectorizer.print_output()
```

```python
class SentenceEmbeddingsVectorizer:
    def __init__(self, corpus):
        self.corpus = corpus
        self.model = SentenceTransformer('all-MiniLM-L6-v2')

    def transform(self):
        sentence_embeddings = self.model.encode(self.corpus)
        return sentence_embeddings

    def print_output(self):
        embeddings = self.transform()
        print("\n--- Sentence Embeddings ---")
        for idx, emb in enumerate(embeddings):
            print(f"Sentence {idx+1} Embedding (first 5 dims): {emb[:5]}...")
sentence_embeddings_vectorizer = SentenceEmbeddingsVectorizer(sample_corpus)
sentence_embeddings_vectorizer.print_output()
```

# **POS Tagging HMM**

```python
import nltk
from nltk.corpus import treebank
from nltk.tag import hmm
from nltk.tokenize import word_tokenize

# Download the 'treebank' data package
nltk.download('treebank')

train_data=treebank.tagged_sents()[:3000]
trainer=hmm.HiddenMarkovModelTrainer()
tagger=trainer.train_supervised(train_data)

sample_sentence="Ram is gifting a book to Sita"

tokens=word_tokenize(sample_sentence)
tagged_tokens=tagger.tag(tokens)
print("Tagged sentence : ",tagged_tokens)
```

# **Syntax analysis**

```python
import nltk
from nltk import CFG
from nltk.parse import ChartParser


nltk.download('punkt')


cfg = CFG.fromstring("""
    S -> NP VP
    NP -> Det N
    VP -> V NP
    Det -> 'the'
    N -> 'cat' | 'dog'
    V -> 'chased' | 'caught'
""")

parser = ChartParser(cfg)


sentence = nltk.word_tokenize("the cat chased the dog")

for tree in parser.parse(sentence):

    tree.pretty_print()
```
