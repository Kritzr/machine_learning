```python
import pandas as pd
file_path = 'C:/Users/2022503035/Documents/machine_learning_2022503035/introduction/dataset.xlsx'
df=pd.read_excel(file_path)
print(df)
```

```python
df['Average marks'] = df[['M1','M2','M3','M4','M5']].mean(axis=1)
print(df)
df.to_excel(file_path, index=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
for index, row in df.iterrows():
    student_name = row['NAME']
    marks = row[['M1','M2','M3','M4','M5']]
    plt.plot(['M1','M2','M3','M4','M5'],marks, marker='o',label=student_name)
    plt.xlabel('SUBJECTS')
    plt.ylabel('MARKS')
    plt.title('MARK DISTRIBUTION')
    plt.legend(title = 'STUDENT')
    plt.grid(True)
    plt.show()
```

```python

def assign_grade(average):
    if average >= 90:
        return 'O' 
    elif average >= 80:
        return 'A+'
    elif average >= 70:
        return 'A'
    elif average >= 60:
        return 'B+'
    elif average >= 50:
        return 'B'
    else:
        return 'F'
df['Grade'] = df['Average Marks'].apply(assign_grade)
print(df)
```
